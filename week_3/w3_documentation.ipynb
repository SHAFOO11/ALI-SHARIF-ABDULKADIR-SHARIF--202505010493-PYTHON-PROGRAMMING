## ACTIVITY 1: Identify the Components

### 1.1 What are the Inputs?

The inputs are the data that the user types into the program:

| # | Input | Description |
|---|-------|-------------|
| 1 | `quiz_1` | The mark for Quiz 1, entered by the user |
| 2 | `quiz_2` | The mark for Quiz 2, entered by the user |
| 3 | `quiz_3` | The mark for Quiz 3, entered by the user |
| 4 | `choice` | The user types **Y** to continue or N to stop |

### 1.2 What is the Process?

The process is what the program **does** with the inputs:

1. **Read** three quiz marks from the user (quiz_1, quiz_2, quiz_3)
2. **Calculate** the average: `average = (quiz_1 + quiz_2 + quiz_3) / 3`
3. **Display** the calculated average to the user
4. **Compare** the average to 50 to decide Pass or Fail
5. **Ask** the user if they want to enter another student's marks (Y/N)
6. **Repeat** from step 1 if user enters Y, or **stop** if they enter N

### 1.3 What is the Output?

The output is what the program **shows** to the user:

| # | Output | When it appears |
|---|--------|-----------------|
| 1 | The average mark | After the three quiz marks are entered |
| 2 | `Pass` | If the average is **50 or more** |
| 3 | `Fail` | If the average is **less than 50** |
| 4 | `Program Ended` | When the user chooses to stop (enters N) |

## ACTIVITY 2: Design the Algorithm

### 2.1 Pseudocode

Pseudocode is writing the steps of the program in simple English before writing real code.

SET choice = "y"

WHILE choice == "y":

    INPUT quiz_1  (ask user: "Enter Quiz 1 mark")
    INPUT quiz_2  (ask user: "Enter Quiz 2 mark")
    INPUT quiz_3  (ask user: "Enter Quiz 3 mark")

    CALCULATE average = (quiz_1 + quiz_2 + quiz_3) / 3

    DISPLAY average

    IF average >= 50:
        DISPLAY "Pass"
    ELSE:
        DISPLAY "Fail"

    INPUT choice  (ask user: "Continue? Select Y/N")

DISPLAY "Program Ended"

### 2.2.1 Sequence

**Sequence** means the program runs steps **one after another, in order**.

**Example from our program:**
```python
quiz_1 = float(input("Enter Quiz 1 mark: "))   # Step 1
quiz_2 = float(input("Enter Quiz 2 mark: "))   # Step 2
quiz_3 = float(input("Enter Quiz 3 mark: "))   # Step 3
average = (quiz_1 + quiz_2 + quiz_3) / 3       # Step 4
print("Average mark:", average)                # Step 5
```
These lines run **top to bottom**, one by one. That is sequence.

### 2.2.2 Selection

**Selection** means the program **makes a decision** and chooses one path. It uses `if` and `else`.

**Example from our program:**
```python
if average >= 50:
    print("Pass")   # Runs ONLY if average is 50 or more
else:
    print("Fail")   # Runs ONLY if average is less than 50
```
- If **YES** → prints `Pass`  
- If **NO** → prints `Fail`  

Only **one** path runs each time. That is selection.

### 2.2.3 Iteration

**Iteration** means the program **repeats** a block of code again and again. It uses a `while` loop.

**Example from our program:**
```python
while choice == "y":
    
    choice = input("Continue? Select Y/N: ")
```
The loop keeps going as long as `choice` equals `"y"`.  
When the user types anything other than `y`, the loop **stops**.  

That is iteration — doing the same thing multiple times.
